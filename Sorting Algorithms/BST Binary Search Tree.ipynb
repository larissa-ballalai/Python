**102. Binary Tree Level Order Traversal**

Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).

Input: root = [3,9,20,null,null,15,7]

Output: [[3],[9,20],[15,7]]

https://leetcode.com/problems/binary-tree-level-order-traversal/



```
# Print level order traversal using Queue

# A node structure
class Node:
    # A utility function to create a new node
    def __init__(self ,key):
        self.data = key
        self.left = None
        self.right = None

# Adding values
root = Node(3)
root.left = Node(9)
root.right = Node(20)
root.right.left = Node(15)
root.right.right = Node(7)

# Iterative Method to print the height of a binary tree
def printLevelOrder(root):
    if root is None:
        return

    result = []
     
    #1. Create an empty queue for level order traversal
    queue = []

    #1.1 Add the root
    queue.append(root)
            
    while len(queue) > 0:
       
        #2. Item which receives the nodes by level
        levelItems = []

        for i in range(len(queue)):

            #3. Pop the current value. If I had here a print, I can print all elements one by one.
            node = queue.pop(0)
            levelItems.append(node.data)
            
            #3. Append right and left values.
            if node.left is not None:
                queue.append(node.left)
    
            if node.right is not None:
                queue.append(node.right)
    
        #2.1 Add the leafs of one level to result
        result.append(levelItems)

    print(result)


print ("Level traversal: ")
printLevelOrder(root)
```

Input: root = [3,9,20,null,null,15,7]

Output: [[3],[20,9],[15,7]]


```
# Print level order traversal using Queue

# A node structure
class Node:
    def __init__(self ,key):
        self.data = key
        self.child = []

# Adding values
root = Node(3)
root.child.append(Node(9))
root.child.append(Node(20))
root.child[0].child.append(Node(15))
root.child[0].child.append(Node(17))

# Iterative Method to print the height of a binary tree
def printLevelOrder(root):
    if root is None:
        return

    result = []
     
    #1. Create an empty queue for level order traversal
    queue = []

    #1.1 Add the root
    queue.append(root)

    #! zigzag
    ltr = True

    while len(queue) > 0:
       
        print(ltr)

        #2. Item which receives the nodes by level
        levelItems = []

        for i in range(len(queue)):

            #3. Pop the current value. If I had here a print, I can print all elements one by one.
            node = queue.pop(0)
            levelItems.append(node.data)
            
            #3. Append the children
            for j in range(len(node.child)) :
              queue.append(node.child[j])
           
        #2.1 Add the leafs of one level to result
        if ltr == True :
          result.append(levelItems)
        else :
          result.append(levelItems[::-1])
        
        ltr = not ltr

    print(result)
    

print ("Level traversal: ")
printLevelOrder(root)
```

Level Order Traversal Of A Binary Tree



Given a binary tree, list the label of the nodes, level by level from left to right. 





```
class TreeNode():
  def __init__(self, label=None, left_ptr=None, right_ptr=None):
      self.label = label
      self.left_ptr = left_ptr
      self.right_ptr = right_ptr

root = TreeNode(2)
root.left_ptr  = TreeNode(5)
root.right_ptr = TreeNode(4)
root.left_ptr.left_ptr   = TreeNode(0)
root.left_ptr.right_ptr  = TreeNode(1)
root.right_ptr.left_ptr  = TreeNode(3)
root.right_ptr.right_ptr = TreeNode(7)

# Complete this function.
# Args: root (TreeNode): Root of the input tree
# Returns: A two dimentional array of integers
```

```
def level_order_traversal(root):
  queue = [root]
  result = []

  while len(queue) > 0:
    size = len(queue)
    level = []

    while size > 0:
      node = queue.pop(0) 
      level.append(node.label)     

      if node.left_ptr:
        queue.append(node.left_ptr)
      if node.right_ptr:
        queue.append(node.right_ptr)
            
      size = size - 1
    
    result.append(level) 
     
  return result
```
