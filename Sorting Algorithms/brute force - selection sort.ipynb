## Selection Sort


```python
def SelectionSort (values):
  for i in range(len(values)):
    min_idx = i

    for j in range(i+1, len(values)):
      if values[min_idx] > values[j]:
        min_idx = j

    values[i], values[min_idx] = values[min_idx], values[i]
    print("tentative", i, " - array:", values) 
```

```python
array = [10, 9, 8, 7, 6]
SelectionSort(array)
```
