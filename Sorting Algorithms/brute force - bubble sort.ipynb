# Bubble Sort

```python
def Bubblesort(values):
  for i in range(len(values)):    
    for j in range(len(values)-i-1):
       if values[j] > values[j+1]:
         values[j], values[j+1] = values[j+1], values[j]

    print("tentative", i, " - array:", values) 
```

```python
array = [10, 9, 8, 7, 6]
Bubblesort(array)
```

! the second for, you add "i" there, because the biggest one will be the last, so you dont need to go there again.
